```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from glob import glob

import librosa
import librosa.display
import IPython.display as ipd

#TO MAKE PLOTS LOOK NICER
from itertools import cycle
sns.set_theme(style="white", palette=None)
color_pal= plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle= cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
data_files=glob('DATA/*.m4a')
```

## visualization

### BIRD1_ ALCON

```python
audio=0
y1, sr1 = librosa.load(data_files[audio])
ipd.Audio(data_files[audio])
```

```python
import math
TAM=1
n1=math.floor(y1.size/TAM)

pd.Series(y1[:n1]).plot(figsize=(10,5), lw=1, title='2 Flautas', color=color_pal[0])
```

```python
y=y1[:n1]

D = librosa.stft(y)
S_db= librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
fig, ax = plt.subplots(figsize=(10,5))
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax)
ax.set_title('STFT SPECTOGRAM', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
```

```python
S=librosa.feature.melspectrogram(y, sr=sr1, n_mels=130,)
S_db_mel= librosa.amplitude_to_db(S, ref=np.max)

fig, ax = plt.subplots(figsize=(10,5))
img = librosa.display.specshow(S_db_mel, x_axis='time', y_axis='log', ax=ax)
ax.set_title('MEL ESPECTOGRAM', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
```
