```python
import matplotlib.pyplot as plt
import numpy as np
import wave

spf = wave.open("lenta.wav", "r")

# Extract Raw Audio from Wav File
signal = spf.readframes(-1)
signal = np.fromstring(signal, dtype=np.uint16)

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(signal)
plt.show()
```
