```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[16,12]
plt.rcParams.update({'font.size':18})
```

```python
dt = 0.001
t = np.arange(0,1,dt)
```

```python
f1=np.sin(2*np.pi*50*t)
f2=np.sin(2*np.pi*120*t)
plt.plot(f2)
```

```python

f = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t) #Suma de 2 frecuencias
plt.plot(f)
```

```python
f_clean = f

f= f + 2.5* np.random.randn(len(t)) # Add some noice
plt.plot(f)
```

```python
plt.plot(t,f,color='c')
plt.xlim(t[0],t[-1])
```

```python
plt.plot(t,f_clean,color='k')
plt.xlim(t[0],t[-1])
```

```python
plt.plot(t,f,color='c', label='Noisy')
plt.plot(t,f_clean,color='k',label='Clean')
plt.xlim(t[0],t[-1])
```

## We are now gonna INFER the original signal using FFT

```python
#Compute the Fast Fourier Transform (FFT)
n =len(t)
fhat= np.fft.fft(f,n) #Fourier coefficients: complex values FOurier coefficients: What are the MAGNITUD and the PHASE of the sin and cosine components of INCREASSING FRECUANCY that you would have to add up (+) to get the data set 'f'
# the vector goes from low frecuency to high frecuency
```

```python
#Compute the Power spectrum PSD
PSD =fhat * np.conj(fhat)/n 

#remember that the conjugate of a complex number multiplied by
# the complex numbers equals the magnitud of the complex number squared

#A vector of powers at each frecuncy

#PSD has units of power: So we are computing the POWER of SPECTRAL DENSITY
```

```python
freq = (1/(dt*n)) * np.arange(n) 
```

```python
L = np.arange(1,np.floor(n/2), dtype='int') #Only plt the first half
```

```python
fig, axs= plt.subplots(2,1)

plt.sca(axs[0])
plt.plot(t,f,color='c', label='Noisy')
plt.plot(t,f_clean, color='k', label='Clean')
plt.xlim(t[0], t[-1])
plt.legend

plt.sca(axs[1])
plt.plot(freq[L],PSD[L],color='c', label='Noisy')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.xlabel('Frecuency in HZ')
plt.ylabel('Power Spectrum')
plt.legend
```

Even thoug the signal is noisy the Power Spectrum has 2 peaks, 1 at 50Hz and another at 120Hz.. That tells you taht most of the power is at those frecuencies and then there is a bunch of noisy

# Filter the signal and denoising

You can actually filter that date selecting a value and say any Fourier coefficient that is lower/higger than that specifi value, We are just gonna set it to Zero, and gonna keep ay Fourier coefficient that es higger/lower that a value.

After Im gonna INVERSE fourier transform and try to get a clean signal

```python
# Use teh PSD to filter out noise
indices = PSD >100 #find all frecuencies with  large power
PSDclean = PSD * indices #Zero out all others
fhat = indices * fhat # Zero out small Fourier coeffs in Y


ffilt = np.fft.ifft(fhat)  #Inverse FFT for filtered time signal
```

```python
fig, axs = plt.subplots(3,1)

plt.sca(axs[0])
plt.plot(t,f,color='c', label='Noisy')
plt.plot(t,f_clean, color='k', label='Clean')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(t,ffilt,color='k', label='Filtered')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[2])
plt.plot(freq[L],PSD[L],color='c', label='Noisy')
plt.plot(freq[L],PSDclean[L], color='k', label='Filtered')
plt.xlim(freq[L[0]],freq[L[-1]])
plt.legend()
```
